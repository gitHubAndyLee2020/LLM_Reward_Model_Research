```python
import os
import numpy as np

def add_column(df_train, folder_name, file_suffix, column_name):
    # Initialize a list to hold the content of files
    file_contents = []

    # Loop through numbers from 0 to 99 to access each file
    for i in range(100):
        file_name = f"{folder_name}/{i}{file_suffix}.txt"
        if os.path.exists(file_name):
            with open(file_name, 'r') as file:
                # Handles end-of-sequence token output from OpenOrca model
                content = file.read().replace('<|im_end|>', '')  # Filter out the <im_end> token
                file_contents.append(content)
        else:
            # If a file does not exist, add a None or an empty string to keep the list size consistent
            file_contents.append(None)

    # Convert the list to a numpy array
    instruct_model_responses = np.array(file_contents)

    # Add the numpy array to df_train as a new column
    df_train[column_name] = instruct_model_responses
```

```python
add_column(df_train_first100, 'instruct_answers_0_to_100', '_mistral_7b_instruct_v0_1', 'instruct_model_response')
add_column(df_train_first100, 'openorca_answers_0_to_100', '_mistral_7b_openorca', 'openorca_model_response')
```

```python
import os
from dotenv import load_dotenv

# Load .env file
load_dotenv('.env')

# Extract the API key
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
import openai

openai.api_key = OPENAI_API_KEY
```

```python
def get_msg(system_prompt, question, response_a, response_b, response_c):
    system_msg = """
You are an assistant tasked with ranking the responses from various Large Language Models based on their human-like qualities. Your goal is to determine which response feels the most natural and intuitive to a human reader."""

    user_msg = f"""
Instruction:
- Rank the responses below based on which one sounds the most like a genuine human response to the provided prompt.
- Rank the responses from best to worst (best is leftmost, worst is rightmost).
- Consider factors such as clarity, relevance, and natural language flow.
- Output the ranking using the labels of each response: A, B, and C. List them in order, separated by commas without spaces. For example: A,B,C or B,C,A.
- Do not output anything else.
[End of Instruction]

Prompt:
System Prompt:
{system_prompt}
Question:
{question}
[End of Prompt]

Responses:
A:
{response_a}
B:
{response_b}
C:
{response_c}
[End of Responses]
"""

    return system_msg, user_msg
```

```python
def get_response(system_msg, user_msg):
  response = openai.ChatCompletion.create(
              model="gpt-3.5-turbo",
              messages=[{"role": "system", "content": system_msg},
                        {"role": "user", "content": user_msg}
              ])

  return response["choices"][0]["message"]["content"]
```

```python
START_IDX = 0
END_IDX = 100
```

```python
def get_ranking(df_train, START_IDX, END_IDX):
  responses = []

  for i in range(START_IDX, END_IDX):
    system_prompt = df_train.loc[i, 'system_prompt']
    question = df_train.loc[i, 'question']
    response_a = df_train.loc[i, 'response']
    response_b = df_train.loc[i, 'instruct_model_response']
    response_c = df_train.loc[i, 'openorca_model_response']
    system_msg, user_msg = get_msg(system_prompt, question, response_a, response_b, response_c)
    response = get_response(system_msg, user_msg)
    responses.append(response)

  return responses
```

```python
rankings = get_ranking(df_train_first100, START_IDX, END_IDX)
```

```python
# A: response, B: instruct_model_response, C: openorca_model_response
rankings
```

```python
import numpy as np

def convert_rankings(data):
    ranking_a = []
    ranking_b = []
    ranking_c = []

    for item in data:
        ranks = item.replace(' ', '').split(',')

        ranking_a.append(ranks.index('A') + 1)
        ranking_b.append(ranks.index('B') + 1)
        ranking_c.append(ranks.index('C') + 1)

    return np.array(ranking_a), np.array(ranking_b), np.array(ranking_c)
```

```python
ranking_a, ranking_b, ranking_c = convert_rankings(rankings)
```

```python
ranking_a, ranking_b, ranking_c
```

```python
# Calculate the average values
avg_ranking_a = np.mean(ranking_a)
avg_ranking_b = np.mean(ranking_b)
avg_ranking_c = np.mean(ranking_c)

# Store them in a dictionary
avg_values = {
    'ranking_a': avg_ranking_a,
    'ranking_b': avg_ranking_b,
    'ranking_c': avg_ranking_c
}

# Sort the dictionary by values
sorted_avg_values = sorted(avg_values.items(), key=lambda x: x[1])

# Print the sorted averages
for key, value in sorted_avg_values:
    print(f"{key}: {value}")
```

```python
df_train_first100['response_rank'] = ranking_a
df_train_first100['instruct_model_response_rank'] = ranking_b
df_train_first100['openorca_model_response_rank'] = ranking_c
```
